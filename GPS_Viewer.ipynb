# GPS Viewer/explorer

This notebook creates an interactive GPS viewer for picking gradients.  Data are from the GeoNet FITS service.

You can pick up to two points per component. The positions of these points will be displayed in the
table on the right of the graphs. You can zoom into the plots and pan around as you wish.

If you click a third point, the first point that you picked will be removed.

You can delete a pick by clicking the point of that pick and hitting the backspace key.

```python
from bokeh.plotting import output_notebook
from helpers.gps_viewer import GPSViewer
from datetime import datetime as dt

output_notebook()
viewer = GPSViewer(
    reciever="LYTT", starttime=dt(2000, 1, 1), endtime=dt(2020, 1, 1))
viewer.plot(plot_width=700)
```
